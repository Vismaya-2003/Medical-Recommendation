Splitting the dataframe "train" into dependent (X) and independent (Y) variables

```python
# Dropping the "prognosis" column from the dataframe to create dependent variable (X)
X = train.drop('prognosis', axis=1)
```

```python
# Displaying the values of the dependent variables (X)
X
```

```python
# Using the dropped "prognosis" column the create independent variable (Y)
Y = train['prognosis']
```

```python
# Displaying the values of the independent variables (Y)
Y
```

Performing Label Encoding on the independent variable (Y)

```python
# Intializing the label encoder
le = LabelEncoder()
```

```python
# Fitting the label encoder on the independent variable (Y)
le.fit(Y)
```

```python
# Transforming the independent variable using label encoder
Y = le.transform(Y)
```

Splitting the dataframe into training and testing datasets

```python
# Splitting the dataframe into training and testing datasets
X_train, X_test, Y_train,Y_test = train_test_split(X,Y, test_size=0.3, random_state=20)
```

```python
# Displaying the shape of the training datasets
print("Training dataset X_train shape:", X_train.shape)
print("Training dataset Y_train shape:", Y_train.shape)
```

```python
# Displaying the shape of the testing datasets
print("Training dataset X_test shape:", X_test.shape)
print("Training dataset Y_test shape:", Y_test.shape)
```

Training the top 5 models

```python
# Create a dictionary to store the models
models = {
    "SVC" : SVC(kernel = 'linear'),
    "Random Forest" : RandomForestClassifier(n_estimators = 100, random_state = 42),
    "Gradient Boosting" : GradientBoostingClassifier(n_estimators = 100, random_state = 42),
    "KNeighbors" : KNeighborsClassifier(n_neighbors = 5),
    "MultinomialNB" : MultinomialNB()
}

for model_name, model in models.items():
     print(model_name, ":", model)
     
     # Training the model
     model.fit(X_train, Y_train)

     # Testing the model
     prediction = model.predict(X_test)

     # Calculating accuracy of the model
     accuracy = accuracy_score(Y_test, prediction)

     # Calculating the confusion matrix
     cm = confusion_matrix(Y_test, prediction)

     print(f"{model_name} accuracy: {accuracy}")
     print(f"{model_name} confusion matrix :")
     print(np.array2string(cm, separator = ','))
```
